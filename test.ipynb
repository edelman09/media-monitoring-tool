```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
import time

# --- Replace with your credentials ---
EMAIL = "Kunal.arora@edelman.com"
PASSWORD = "dashboard1"
TARGET_TOPIC = "MS Work Trend Index"
TARGET_CATEGORY = "Ungrouped"

# --- Setup Chrome Driver ---
options = Options()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    # Login
    driver.get("https://app.talkwalker.com/app/login")
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, "email"))).send_keys(EMAIL)
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "next-button"))).click()
    WebDriverWait(driver, 15).until(EC.visibility_of_element_located((By.NAME, "password"))).send_keys(PASSWORD)
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "login-button"))).click()

    # Click "Topic Analytics"
    topic_analytics_card = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "//div[contains(@class, 'descriptive-card-title')]//span[contains(text(), 'Topic Analytics')]/ancestor::div[contains(@class, 'descriptive-card')]"))
    )
    topic_analytics_card.click()
    print("✅ Clicked on 'Topic Analytics'")
    # time.sleep(5)

    # Wait for sidebar categories
    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "view-container")))
    time.sleep(2)

    category_containers = driver.find_elements(By.XPATH, "//div[contains(@class,'p-sbtm-item') and contains(@class, 'group')]")
    all_topics = {}
    ungrouped_toggle = None
    ungrouped_topics = []

    for category in category_containers:
        try:
            category_name = category.find_element(By.XPATH, ".//span[contains(@class, 'item-label')]").text.strip()
            print(f"\n📁 Category: {category_name}")
            toggle_button = category.find_element(By.XPATH, ".//button[contains(@class,'action-icon')]")
            if len(category.find_elements(By.XPATH, "following-sibling::div[contains(@class, 'child')]")) == 0:
                toggle_button.click()
                time.sleep(1)
            topics = category.find_elements(By.XPATH, "following-sibling::div[contains(@class, 'child')]")
            topic_list = []
            for topic in topics:
                try:
                    topic_name = topic.find_element(By.XPATH, ".//span[contains(@class, 'item-label')]").text.strip()
                    print(f"   • {topic_name}")
                    topic_list.append((topic_name, topic))
                except:
                    continue
            all_topics[category_name] = topic_list
            if category_name == TARGET_CATEGORY:
                ungrouped_toggle = toggle_button
                ungrouped_topics = topic_list
        except Exception as e:
            print(f"⚠️ Skipping category due to error: {e}")
            continue

    print(f"\n🔎 Looking for topic: {TARGET_TOPIC} in {TARGET_CATEGORY}")
    if not ungrouped_topics and ungrouped_toggle:
        print("🔁 Expanding Ungrouped category again...")
        ungrouped_toggle.click()
        # time.sleep(2)
        ungrouped_topics = category.find_elements(By.XPATH, "following-sibling::div[contains(@class, 'child')]")

    # Click on the specific topic
    found = False
    for topic in ungrouped_topics:
        try:
            topic_name = topic.find_element(By.XPATH, ".//span[contains(@class, 'item-label')]").text.strip()
            if topic_name == TARGET_TOPIC:
                topic.click()
                print(f"✅ Clicked on topic: {topic_name}")
                found = True
                break
        except:
            continue

    if not found:
        print("❌ Topic not found in Ungrouped.")
    # time.sleep(1)

    # Click "Results" tab icon from sidebar
    results_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "//i[contains(@class,'tw3-icon-results-list')]"))
    )
    results_icon.click()
    print("✅ Navigated to Results tab")
    # time.sleep(3)

    # Step: Hover over the correct header to trigger the correct menu
    print("🕵️ Hovering over the widget header containing 'Group' and 'Sort by'...")
    results_widget = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((
            By.XPATH,
            "//div[contains(@class,'widget-card-section') and .//div[contains(text(),'Group') and contains(@class,'bar-dd-menu-header')] and .//div[contains(text(),'Sort by') and contains(@class,'bar-dd-menu-header')]]"
        ))
    )

    # Hover to ensure menu activation
    actions = ActionChains(driver)
    actions.move_to_element(results_widget).perform()
    # time.sleep(2)

    # Locate the correct three-dot menu inside the matched results widget
    print("🧭 Looking for the correct three-dot menu inside the results widget...")
    three_dot_button = results_widget.find_element(
        By.XPATH,
        ".//div[contains(@class,'item-header-action-icon')]//button[i[contains(@class,'tw3-icon-three-dots')]]"
    )

    # Click the three-dot menu
    three_dot_button.click()
    print("✅ Correct export options menu opened")
    # time.sleep(2)

    # Click on Export in dropdown
    export_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//div[@id='EXPORT' and contains(@class,'clickable')]"))
    )
    export_button.click()
    print("📦 Export triggered successfully.")
    # time.sleep(5)

    # Step: Wait for the export modal to appear
    print("📦 Waiting for export modal...")
    export_modal = WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.XPATH, "//div[@role='dialog' and contains(@class,'modal-bubble')]"))
    )

    # Step: Click the "CSV" format tile
    print("📄 Selecting CSV option...")
    csv_button = export_modal.find_element(
        By.XPATH,
        ".//div[contains(@class,'custom-button-tab-label') and text()='CSV']/parent::div[@tabindex='0']"
    )
    csv_button.click()
    print("✅ CSV selected")

    # Step: Wait for and click the active 'Export' button
    print("🚀 Waiting for Export button to become enabled...")
    export_button = WebDriverWait(driver, 15).until(
        EC.element_to_be_clickable((By.ID, "model-confirm-button"))
    )
    export_button.click()
    print("📤 Export initiated")

    # Step: Wait for the confirmation and download link
    print("📥 Waiting for download link...")
    download_link = WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((
                By.XPATH,
                "//div[contains(@class,'notification-renderer-item')]//div[contains(@class,'info-title')]//a[contains(@href,'.csv') and contains(text(),'here')]"
            ))
        )
    download_url = download_link.get_attribute("href")
    print(f"✅ CSV file ready: {download_url}")
    time.sleep(2)

    # Step 5: Click the anchor link via JavaScript
    # driver.execute_script("arguments[0].click();", download_link)
    driver.get(download_url)
    print("📁 CSV file download triggered by JavaScript click")
    time.sleep(5)

except Exception as e:
    print("❌ An error occurred:", e)

finally:
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
import time

# --- Replace with your credentials ---
EMAIL = "Kunal.arora@edelman.com"
PASSWORD = "dashboard1"

# --- Setup Chrome Driver ---
options = Options()
options.add_argument("--start-maximized")
# To run in headless mode (without opening a browser window), uncomment the next line
# options.add_argument("--headless")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

try:
    wait = WebDriverWait(driver, 20)
    actions = ActionChains(driver)

    # Step 1: Login
    driver.get("https://app.talkwalker.com/app/login")
    wait.until(EC.presence_of_element_located((By.NAME, "email"))).send_keys(EMAIL)
    wait.until(EC.element_to_be_clickable((By.ID, "next-button"))).click()
    wait.until(EC.visibility_of_element_located((By.NAME, "password"))).send_keys(PASSWORD)
    wait.until(EC.element_to_be_clickable((By.ID, "login-button"))).click()
    print("✅ Logged into Talkwalker.")

    # Step 2: Click on the project dropdown
    print("📂 Waiting for Project Selection button...")
    project_dropdown_button = wait.until(EC.element_to_be_clickable((
        By.XPATH, "//button[contains(@class,'p-navbar-project-selection')]"
    )))
    project_dropdown_button.click()
    print("✅ Project dropdown opened.")

    # Step 3: Fetch all projects
    print("🔎 Fetching all available projects...")
    project_elements = wait.until(EC.presence_of_all_elements_located((
        By.XPATH, "//div[contains(@class,'nav-menu-body-content')]//div[contains(@class,'navbar-project') and contains(@class,'p-menu-item')]//div[contains(@class,'menu-label')]"
    )))

    project_names = []
    for elem in project_elements:
        project_names.append(elem.text.strip())

    print("\n📋 Available Projects:")
    for idx, name in enumerate(project_names, start=1):
        print(f"{idx}. {name}")

    # Step 4: Ask user for project selection
    project_choice = input("\nEnter the project number you want to select: ").strip()
    project_choice_idx = int(project_choice) - 1

    if project_choice_idx < 0 or project_choice_idx >= len(project_names):
        raise ValueError("❌ Invalid project number selected.")

    selected_project_name = project_names[project_choice_idx]
    print(f"✅ Selecting project: {selected_project_name}")

    # Step 5: Click the selected project
    selected_project_element = project_elements[project_choice_idx]
    driver.execute_script("arguments[0].click();", selected_project_element)

    # Wait for project switching
    time.sleep(3)

    # Step 6: Proceed to Topic Analytics first
    print("\n🚀 Navigating to 'Topic Analytics'...")
    topic_analytics_card = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "//div[contains(@class, 'descriptive-card-title')]//span[contains(text(), 'Topic Analytics')]/ancestor::div[contains(@class, 'descriptive-card')]"))
    )
    topic_analytics_card.click()
    print("✅ Clicked on 'Topic Analytics'")

    WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "view-container")))
    time.sleep(2)

    # 📅 Step 7: NOW Ask for Time Period Selection
    print("\n🕒 Select Time Period:")
    time_options = {
        "1": {"id": "d1", "label": "1D"},
        "2": {"id": "d7", "label": "7D"},
        "3": {"id": "d30", "label": "30D"},
        "4": {"id": "m3", "label": "3M"},
        "5": {"id": "m6", "label": "6M"},
        "6": {"id": "y1", "label": "1Y"},
    }

    for key, value in time_options.items():
        print(f"{key}. {value['label']}")

    time_choice = input("\nEnter the number corresponding to the time period: ").strip()

    try:
        if time_choice in time_options:
            selected_option = time_options[time_choice]
            data_id = selected_option["id"]
            label = selected_option["label"]
            
            xpath = f"//div[@data-id='{data_id}']"
            element = wait.until(EC.element_to_be_clickable((By.XPATH, xpath)))
            element.click()
            print(f"✅ Selected Time Period: {label}")
        else:
            print("⚠️ Invalid time period selection. Proceeding with default.")
    except Exception as e:
        print(f"⚠️ Error while selecting time period: {e}")


    # Step 8: Show categories, ask user to select category first
    print("\n📂 Fetching available categories...")
    time.sleep(2)
    category_containers = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[contains(@class,'p-sbtm-item') and contains(@class, 'group')]")))
    all_topics = {}
    category_names = []

    for idx, category in enumerate(category_containers, start=1):
        try:
            category_name = category.find_element(By.XPATH, ".//span[contains(@class, 'item-label')]").text.strip()
            print(f"{idx}. {category_name}")
            all_topics[category_name] = category
            category_names.append(category_name)
        except Exception as e:
            print(f"⚠️ Skipping category due to error: {e}")
            continue

    category_choice = input("\nEnter the number of the category you want to select: ").strip()
    selected_category_idx = int(category_choice) - 1

    if selected_category_idx < 0 or selected_category_idx >= len(category_names):
        raise ValueError("❌ Invalid category number selected.")

    selected_category_name = category_names[selected_category_idx]
    selected_category_element = all_topics[selected_category_name]

    print(f"\n✅ Selected Category: {selected_category_name}")

    # Expand selected category if it's not already expanded
    try:
        # Check if the expand icon is a right-facing triangle, which means it's collapsed
        toggle_button = selected_category_element.find_element(By.XPATH, ".//i[contains(@class, 'tw3-icon-triangle-right')]")
        # If the button is found, it means the category is collapsed, so click it
        toggle_button.click()
        time.sleep(1) # Wait for animation
        print(f"📖 Expanded '{selected_category_name}' to show topics.")
    except Exception:
        # If the right-triangle icon is not found, it's likely already expanded or has a different icon
        print(f"📖 Category '{selected_category_name}' is already expanded.")


    # Now fetch topics inside selected category
    topics_in_category = selected_category_element.find_elements(By.XPATH, "following-sibling::div[contains(@class, 'child')]")

    topic_names = []
    for idx, topic in enumerate(topics_in_category, start=1):
        try:
            topic_name = topic.find_element(By.XPATH, ".//span[contains(@class, 'item-label')]").text.strip()
            print(f"{idx}. {topic_name}")
            topic_names.append((topic_name, topic))
        except:
            continue
    
    if not topic_names:
        raise Exception(f"❌ No topics found under the category '{selected_category_name}'.")

    topic_choice = input("\nEnter the number of the topic you want to select: ").strip()
    selected_topic_idx = int(topic_choice) - 1

    if selected_topic_idx < 0 or selected_topic_idx >= len(topic_names):
        raise ValueError("❌ Invalid topic number selected.")

    selected_topic_name, selected_topic_element = topic_names[selected_topic_idx]

    print(f"✅ Selected Topic: {selected_topic_name}")

    # --- THIS IS THE CORRECTED PART ---
    # Click the 'a' tag within the topic element, which is the actual clickable link
    topic_clickable_part = selected_topic_element.find_element(By.XPATH, ".//a[contains(@class, 'item-container')]")
    driver.execute_script("arguments[0].click();", topic_clickable_part)
    time.sleep(2)

    # Step 9: Proceed to Results tab
    results_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "//i[contains(@class,'tw3-icon-results-list')]"))
    )
    results_icon.click()
    print("✅ Navigated to Results tab")
    
    time.sleep(5)
    print("🕵️ Hovering over the widget header containing 'Group' and 'Sort by'...")
    results_widget = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((
            By.XPATH,
            "//div[contains(@class,'widget-card-section') and .//div[contains(text(),'Group') and contains(@class,'bar-dd-menu-header')] and .//div[contains(text(),'Sort by') and contains(@class,'bar-dd-menu-header')]]"
        ))
    )
    actions.move_to_element(results_widget).perform()

    print("🧭 Looking for the correct three-dot menu inside the results widget...")
    three_dot_button = results_widget.find_element(
        By.XPATH,
        ".//div[contains(@class,'item-header-action-icon')]//button[i[contains(@class,'tw3-icon-three-dots')]]"
    )
    three_dot_button.click()
    print("✅ Correct export options menu opened")

    export_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//div[@id='EXPORT' and contains(@class,'clickable')]"))
    )
    export_button.click()
    print("📦 Export triggered successfully.")

    print("📦 Waiting for export modal...")
    export_modal = WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.XPATH, "//div[@role='dialog' and contains(@class,'modal-bubble')]"))
    )

    print("📄 Selecting CSV option...")
    csv_button = export_modal.find_element(
        By.XPATH,
        ".//div[contains(@class,'custom-button-tab-label') and text()='CSV']/parent::div[@tabindex='0']"
    )
    csv_button.click()
    print("✅ CSV selected")

    print("🚀 Waiting for Export button to become enabled...")
    export_button = WebDriverWait(driver, 15).until(
        EC.element_to_be_clickable((By.ID, "model-confirm-button"))
    )
    export_button.click()
    print("📤 Export initiated")

    print("📥 Waiting for download link...")
    download_link = WebDriverWait(driver, 60).until( # Increased wait time for slow exports
            EC.presence_of_element_located((
                By.XPATH,
                "//div[contains(@class,'notification-renderer-item')]//div[contains(@class,'info-title')]//a[contains(@href,'.csv') and contains(text(),'here')]"
            ))
        )
    download_url = download_link.get_attribute("href")
    print(f"✅ CSV file ready: {download_url}")
    time.sleep(2)

    driver.get(download_url)
    print("📁 CSV file download triggered by opening the URL")
    # Add sufficient time for the download to complete
    time.sleep(10)

except Exception as e:
    print(f"\n❌ An error occurred: {e}")

finally:
    print("\n▶️ Script finished.")
    driver.quit()
```
